```python
# Install (if needed)
# pip install 

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from ta.momentum import RSIIndicator
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
```

```python
# Data Collection using Yahoo Finance
# Purpose

# Download historical stock market data using yfinance.

ticker = "AAPL"

df = yf.download(
    ticker,
    start="2020-01-01",
    end="2026-01-01"
)

df.head()


symbol = 'AAPL'
start = '2024-01-01'
end = '2026-01-01'

df = yf.download(symbol, start=start, end=end)
df.head()
```

```python
# Data Cleaning
# Purpose

# Handle missing values and prepare the dataset.

df = df.dropna()
```

```python
# Moving Average
# Purpose

# Identify long-term and short-term trends.

df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()
```

```python
# RSI Calculation

# Purpose

# Detect overbought and oversold conditions.

from ta.momentum import RSIIndicator

rsi = RSIIndicator(df['Close'], window=14)
df['RSI'] = rsi.rsi()
```

```python
# Momentum Calculation

# Purpose

# Measure price strength and trend continuation.

df["Momentum"] = (
    df["Close"] /
    df["Close"].shift(10)
) - 1

Interpretation:

Positive = Bullish Momentum
Negative = Bearish Momentum
```

```python
# Export Data to CSV
df.to_csv('AAPL_stock_analysis.csv')
print('CSV exported successfully')
```

```python
# Buy / Sell / Hold Signal Generation
# Purpose

# Generate trading recommendations.

def signal(row):

    if (
        row["MA20"] > row["MA50"]
        and row["RSI"] < 70
        and row["Momentum"] > 0
    ):
        return "BUY"

    elif (
        row["MA20"] < row["MA50"]
        or row["RSI"] > 70
    ):
        return "SELL"

    else:
        return "HOLD"

df["Signal"] = df.apply(
    signal,
    axis=1
)
```

```python
# Visualization

plt.figure(figsize=(14,6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA20'], label='MA20')
plt.plot(df['MA50'], label='MA50')
plt.title('Stock Price with Moving Averages')
plt.legend()
plt.show()

plt.figure(figsize=(14,4))
plt.plot(df['RSI'], label='RSI', color='purple')
plt.axhline(70, linestyle='--', color='red')
plt.axhline(30, linestyle='--', color='green')
plt.title('RSI Indicator')
plt.legend()
plt.show()

plt.figure(figsize=(14,6))


plt.plot(df["Close"], label="Close") 
plt.plot(df["MA20"], label="MA20")


plt.plot(df["MA50"], label="MA50") 


plt.legend()
plt.title("Price with Moving Averages")
plt.show()
```
